# Classifiers Performance Analysis

```python
result = pl1.read_data('clf_evaluations.csv', 'csv')
METRICS = ['auc', 'f1_score','precision','recall','accuracy','train_time']

def select_top_20(result, metric):
    top_20 = result.sort(metric, ascending=False).head(20)
    return top_20

for metric in METRICS:
    print(select_top_20(result, metric)[['model_type', metric]])
```

From the above talbes, we can see Decision Tree does best based on the roc_auc_score and recall. Compared to other methods, Ada Boost and Random Forest really take a long time to train.
